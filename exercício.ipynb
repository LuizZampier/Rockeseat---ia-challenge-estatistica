```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Dicionário de faturamento
dict_faturamento = {
    'data_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
        ],
    'valor': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
        ]
}
```

```python
df = pd.DataFrame(dict_faturamento)
```

```python
df
```

```python
df.sort_values('data_ref', inplace=True)
```

## Média

```python
print(df.valor.mean())
```

## Gráfico de Barras

```python
df.plot.bar(x='data_ref', y='valor', color='royalblue')
plt.xlabel('Data')
plt.ylabel('Valor')
plt.title('Valor de Venda por Data')
```

## Gráfico de Linha

```python
df.plot.line(x='data_ref', y='valor', color='royalblue')
plt.xlabel('Data')
plt.ylabel('Valor')
plt.title('Valor de Venda por Data')
plt.xticks(rotation=90)
```
